```python
# Connecting to the db
import lib.handle_db as dbh

# read and write csv files
import lib.handle_csv as csv_rw

# date functions
from datetime import datetime, date, timedelta

#CR libraries
from crossref.restful import Works, Etiquette

# CR etiquette
my_etiquette = Etiquette('UK Catalysis Hub - Catalysis Data Infrastructure', 
                         'Prototype 2', 
                         'https://ukcatalysishub.co.uk/core/', 
                         'nieva@rc-harwell.ac.uk')

# cr api for retrieving works 
works = Works(etiquette=my_etiquette)

# get a list of ids, titles, dois, and links
# for pubs with no data from the app database
def get_pub_app_data():
    search_in = 'articles'
    fields_required = "id, title, doi, link, url, referenced_by_count"
    filter_str = "status = 'Added'"
    db_titles = db_conn.get_values(search_in, fields_required, filter_str)
    return db_titles

def update_citations(pub_id, citation_val):
    db_conn.set_value_table('articles', pub_id, 'referenced_by_count', citation_val)
```

```python
db_conn = dbh.DataBaseAdapter('../mcc_data/development_2024bk.sqlite3')
```

```python
pubs_list = get_pub_app_data()
```

```python
for a_pub in pubs_list:
    pub_id = a_pub[0]
    pub_title = a_pub[1]
    pub_doi = a_pub[2]
    pub_citations = a_pub[5]
    if pub_id > 506:
        print("getting data for doi", pub_doi)
        print("\tTitle:", pub_title)
        print("\tID:", pub_id)
        print("\tCitations:", pub_citations)

        print(works.filter(doi=pub_doi.strip()).url)
    
        cr_works = works.filter(doi=pub_doi.strip())
        for crwk in cr_works: 

            # print(crwk.keys())
            #if "abstract" in crwk.keys():
            #    print(crwk["abstract"])
            if "is-referenced-by-count" in crwk.keys():

                if crwk["is-referenced-by-count"] != pub_citations:
                    print("Update citations to ", crwk["is-referenced-by-count"])
                    update_citations(pub_id, crwk["is-referenced-by-count"])
```
